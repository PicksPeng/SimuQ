```python
from simuq import QSystem, Qubit
```

```python
n=2
qs = QSystem()
qubits=[Qubit(qs) for i in range(2*n)]
h1=0
h2=0
h3=0
for i in range(n):
    h1+=qubits[i].X+qubits[i].Z
    h2+=qubits[i].Y
    h3+=qubits[i+n].Z
h=h1*h3+h2
qs.add_evolution(h, 1)



ionq.compile(qs)
solved_unitary=ionq.prog_obj.to_matrix()
```
